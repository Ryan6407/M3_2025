```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime


def day_of_year(month: int, day: int) -> int:
    return (datetime(2025, month, day) - datetime(2025, 1, 1)).days + 1

def find_cos_of_angle_of_incidence(month, day, hour, minute):
    B14 = -90
    B9prev = -90.049
    n = day_of_year(month, day)
    B51 = 0.613 # radians
    B9 = np.radians(23.45 * np.sin(np.radians((360/365) * (n+284))))
    E22 = hour + minute/60
    D18 = np.radians(360*(n-1)/365)
    B20 = 229.2*(0.000075+(0.001868*np.cos(D18))-(0.032077*np.sin(D18))-(0.014615*np.cos(2*D18))-(0.04089*np.sin(2*D18)))
    D22 = E22+(B20/60)+((4/60)*(B14-B9prev))
    B24 = np.radians(-1 * 15 * (12-D22))
    return ((np.sin(B51))*(np.sin(B9)))+((np.cos(B9))*(np.cos(B24))*(np.cos(B51)))


def find_irradiance(month, day, hour, minute):
    B3 = day_of_year(month, day)
    B4 = (360*(B3-1))/365
    B5 = np.radians(B4)
    D5 = 1367*(1.00011+(0.034221*np.cos(B5))+(0.00128*np.sin(B5))+(0.000719*np.cos(2*B5))+(0.000077*(np.sin(2*B5))))
    D6 = (0.99*(0.4237-(0.00821*((6-0.103)**2))))+((0.99*(0.5055+(0.00595*((6.5-0.103)**2))))*np.exp((-1 * (1.01 * (0.2711 + (0.01858 * ((2.5 - 0.103)**2))))) / find_cos_of_angle_of_incidence(month, day, hour, minute)))
    D8 = D5 * D6
    D9 = D8*find_cos_of_angle_of_incidence(month, day, hour, minute)
    D10 = (0.271-(0.294*D6))*D5*round(find_cos_of_angle_of_incidence(month, day, hour, minute), 3)
    return D10 + D9

def Q_cond(T_out, T_in, k, A_wall, d):
    return k * A_wall * (T_out - T_in) / d

def Q_vent(rho, C_p, V, N_air_changes, T_out, T_in):
    return rho * C_p * V * N_air_changes * (T_out - T_in)

def perturb_params(original_params, percent=5):
    return {k: v * np.random.uniform(1-percent/100, 1+percent/100) 
            for k, v in original_params.items()}

def simulate_house(A_floor, num_people, params, windspeed_data, humidity_data, T_out_data):
    AIR_DENSITY = params['air_density']
    AIR_HEAT_CAPACITY = params['air_heat_capacity']
    DRYWALL_THERMAL_CONDUCTIVITY = params['drywall_thermal_conductivity']
    DRYWALL_THICKNESS = params['drywall_thickness']
    HEIGHT_HOUSE = params['height_house']
    
    V = A_floor * HEIGHT_HOUSE
    A_wall = A_floor * 0.15
    dt = 1
    time_steps = len(windspeed_data)
    
    T_in = np.zeros(time_steps)
    T_in[0] = T_out_data[0]
    
    for t in range(1, time_steps):
        wind_speed = windspeed_data[t]
        T_out = T_out_data[t-1]
        
    
        Q_gain = Q_cond(T_out, T_in[t-1], DRYWALL_THERMAL_CONDUCTIVITY, A_wall, DRYWALL_THICKNESS) + (num_people * 330)
        N_air_changes = 0.1 + 0.4 * np.tanh(0.1 * wind_speed)
        Q_loss = Q_vent(AIR_DENSITY, AIR_HEAT_CAPACITY, V, N_air_changes, T_out, T_in[t-1])
        
     
        T_in[t] = T_in[t-1] + dt * (Q_gain - Q_loss) / (AIR_DENSITY * V * AIR_HEAT_CAPACITY)
        T_in[t] = np.clip(T_in[t], 60, 120)
    
    return T_in



heatwave_data = pd.read_csv("/kaggle/input/d/mantramittal12/m3data-provided/for_all - Heatwave Temps.csv")
dwellings_data = pd.read_csv("/kaggle/input/d/mantramittal12/m3data-provided/for_all - Q1 Dwellings.csv")


original_params = {
    'air_density': 1.293,
    'air_heat_capacity': 1,
    'drywall_thermal_conductivity': 0.2,
    'drywall_thickness': 0.0127,
    'height_house': 2.5
}

base_windspeed = [float(x) for x in heatwave_data["Unnamed: 6"][2:26].values]
base_humidity = [float(x) for x in heatwave_data["Unnamed: 13"][2:26].values]
base_T_out = [(float(x)*9/5)+32 for x in heatwave_data["Unnamed: 2"][2:26].values]


N_TRIALS = 10000
time = np.arange(24)

fig, axs = plt.subplots(2, 2, figsize=(15, 10))
plt.subplots_adjust(hspace=0.4, wspace=0.3)

for dwelling_idx in range(4):
    A_floor = int(dwellings_data.iloc[7].values[1:5][dwelling_idx])
    num_people = int(dwellings_data.iloc[10].values[1:5][dwelling_idx])
    
    all_trials = []
    

    for _ in range(N_TRIALS):
        params = perturb_params(original_params)
        perturb_wind = [x * np.random.uniform(0.95, 1.05) for x in base_windspeed]
        perturb_temp = [x * np.random.uniform(0.95, 1.05) for x in base_T_out]
        trial_result = simulate_house(A_floor, num_people, params, perturb_wind, base_humidity, perturb_temp)
        all_trials.append(trial_result)
    

    mean_temp = np.mean(all_trials, axis=0)
    std_temp = np.std(all_trials, axis=0)
    cv_sum = np.mean(std_temp / np.where(mean_temp != 0, mean_temp, 1e-6))
    print(cv_sum)

    row = dwelling_idx // 2
    col = dwelling_idx % 2
    ax = axs[row, col]
    

    ax.plot(time, mean_temp, label='Mean Indoor Temp', lw=1.5)
    ax.fill_between(time, mean_temp-2 * std_temp, mean_temp+ 2 * std_temp, alpha=0.2, label='2 STD')
    ax.plot(time, base_T_out[:24], '--', label='Outdoor Temp', lw=1)
    
   

    ax.set_title(f'Dwelling {dwelling_idx+1}', fontsize=12)
    ax.set_xlabel('Hour', fontsize=10)
    ax.set_ylabel('Temperature (F)', fontsize=10)
    ax.tick_params(axis='both', labelsize=8)
    ax.grid(True, alpha=0.3)
    ax.legend(fontsize=8)

plt.suptitle('Sensitivity Analysis of Dwelling Temperatures (10,000 Trials)', y=1.02, fontsize=14)
plt.show()
```
